**Scatterplot (Numerical - Numerical)->**

Helps in finding relationship between between two numerical values.

```python
sns.scatterplot(x = tips['total_bill'], y = tips['tip'])
```

```python
sns.scatterplot(x = tips['total_bill'], y = tips['tip'], hue = tips['sex'], style = tips['smoker'], size = tips['size'])
#This is a type of multi-variate data analysis
```

**Barplot (Numerical - Categorical)->**

Helps in finding relationship between numerical and categorical values.

```python
titanic.head()
```

```python
sns.barplot(x = titanic['Pclass'], y = titanic['Fare'])
#Shows average fare in each class
```

```python
sns.barplot(x = titanic['Pclass'], y = titanic['Fare'], hue = titanic['Sex'])
#Multi-variate analysis
```

**Boxplot (Numerical - Categorical)->**

```python
sns.boxplot(x = titanic['Sex'], y = titanic['Age'])
```

```python
sns.boxplot(x = titanic['Sex'], y = titanic['Age'], hue = titanic['Survived'])
```

**Distplot (Numerical - Categorical)->**

```python
sns.distplot(titanic[titanic['Survived'] == 0]['Age'], hist = False, color = 'blue')
sns.distplot(titanic[titanic['Survived'] == 1]['Age'], hist = False, color = 'red')
```

The above plot shows that people with less age have more surviving rate then people with more age

**Heatmap (Categorical - Categorical)->**

Helps in finding relationship between two categorical values.

```python
pd.crosstab(titanic['Pclass'], titanic['Survived'])
```

The above data shows that in a class the number of people survived and not survived.

```python
sns.heatmap(pd.crosstab(titanic['Pclass'], titanic['Survived']))
```

```python
titanic.groupby('Pclass').mean('Survived')['Survived'] * 100 #It is calculating how many people survived in each Pclass in percentage
```

```python
(titanic.groupby('Sex').mean('Survived')['Survived'] * 100).plot(kind = 'bar')
```

```python
flights.head()
```

```python
flights.pivot_table(values = 'passengers', index = 'month', columns = 'year') #It creates a pivot table
```

Since we cannot see a pattern in above table.

So to understand these types of data heatmap is very useful.

```python
sns.heatmap(flights.pivot_table(values = 'passengers', index = 'month', columns = 'year'))
```

**Clustermap (Categorical - Categorical)->**

```python
sns.clustermap(flights.pivot_table(values = 'passengers', index = 'month', columns = 'year')) #It captures hierarichal relation, captured relations between values
```

The above cluster map shows that what all months are related to each other.

Like summer months are related to each other.
