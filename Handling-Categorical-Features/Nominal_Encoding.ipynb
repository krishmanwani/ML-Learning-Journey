**1. One hot Encoding using pandas** (Here we get multi-collinearity)

```python
pd.get_dummies(df, columns=['fuel', 'owner'])
#Here fuel is divided into 4 new columns and owner is divided into 5 new columns
```

**2. K-1 Encoding** (Here multi-collinearity is not a problem)

```python
pd.get_dummies(df, columns=['fuel', 'owner'], drop_first = True)
#One-one category is removed from both fuel and owner
```

Difference between pandas and sklearn is that pandas doesn't remember which column is first and sklearn remembers it
