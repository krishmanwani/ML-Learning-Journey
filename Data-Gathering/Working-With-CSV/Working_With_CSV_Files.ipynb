**Opening a CSV file from an URL**

```python
import requests
from io import StringIO

url = "https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv"
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}
req = requests.get(url, headers=headers)
data = StringIO(req.text)

pd.read_csv(data)
```

**Index_col parameter**

Generally pandas create a new column for index.

If we don't need that we can assign a column as index in our dataset.

```python
pd.read_csv('/content/aug_train.csv', index_col = 'enrollee_id')
```

**Header Parameter**

When the column names are stored in second column then we can set which column we have to be use as the column name

```python
pd.read_csv('/content/test.csv', header = 1)
```

**Use_cols Parameter**

When we want specific columns then we can use this function.

```python
pd.read_csv('/content/aug_train.csv', usecols = ['enrollee_id', 'city', 'gender'])
```

**Squeeze parameters**

```python
pd.read_csv('aug_train.csv',usecols = ['gender']).squeeze()
```

**dtypes Parameter**

**Handling dates**

If we have dates in object(string) format then we can convert it in dates format so that the operations on dates can be used.

```python
pd.read_csv('/content/IPL Matches 2008-2020.csv',parse_dates=['date']).info()
#parse_dates converts the object to dates
```

```python
def rename(name):
  if name == "Royal Challengers Bangalore":
    return "RCB"
  else:
    return name
```

```python
rename("Royal Challengers Bangalore")
```

**Convertors**

Conerts a specific name to another name for all the values in a column by calling a function.
The function should rename the name1 to name2 just like in the above function

```python
pd.read_csv('/content/IPL Matches 2008-2020.csv',converters={'team1':rename})
```

**na_values parameters**

Not all datasets have null values as NaN they can have null values as '-' or anything. For these type of datasets we can set which one we have to treat as  null values

```python
pd.read_csv('/content/aug_train.csv')
```

```python
#In above code the null values are NaN but if want Male as null values then we can write this code
pd.read_csv('/content/aug_train.csv',na_values=['Male',])
```

**Loading a huge dataset**

Sometimes the dataset is very big which our device is not able to handle. So we can break the dataset into chunks.

```python
dfs = pd.read_csv('/content/aug_train.csv',chunksize=5000)
```

```python
for i in dfs:
  print(i.shape)

#When we create chunks the only difference we have to in our codes is that everytime
#we do some operations on the dataset we have to use a loop to iterate on the chunks we created.
```
